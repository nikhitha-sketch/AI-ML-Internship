```python
# AI & ML Internship - Day 3
# Topic : NumPy Fundamentals
# Description : Working with NumPy arrays and mathematical operations
import numpy as np

print("=" * 60)
print("          NUMPY FUNDAMENTALS DEMONSTRATION")
print("=" * 60)

# Create a NumPy array
student_marks = np.array([78, 85, 92, 67, 88])

print("\nOriginal Array:")
print(student_marks)

# Array Information
print("\nArray Information")
print("Dimensions :", student_marks.ndim)
print("Shape      :", student_marks.shape)
print("Data Type  :", student_marks.dtype)
print("Size       :", student_marks.size)

# Indexing
print("\nIndexing")
print("First Element :", student_marks[0])
print("Last Element  :", student_marks[-1])

# Slicing
print("\nSlicing")
print("First Three Elements :", student_marks[:3])
print("Last Two Elements    :", student_marks[-2:])

# Mathematical Operations
print("\nMathematical Operations")
print("Add 5      :", student_marks + 5)
print("Multiply 2 :", student_marks * 2)
print("Square     :", np.square(student_marks))

# Statistical Operations
print("\nStatistical Analysis")
print("Sum     :", np.sum(student_marks))
print("Average :", np.mean(student_marks))
print("Maximum :", np.max(student_marks))
print("Minimum :", np.min(student_marks))
print("Median  :", np.median(student_marks))
print("Standard Deviation :", np.std(student_marks))

# 2D Array
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("\n2D Array")
print(matrix)

print("\nElement at Row 2, Column 3 :", matrix[1, 2])

print("\nRow-wise Sum :", np.sum(matrix, axis=1))
print("Column-wise Sum :", np.sum(matrix, axis=0))

print("\n" + "=" * 60)
print("NumPy Fundamentals Completed Successfully")
print("=" * 60)
```
